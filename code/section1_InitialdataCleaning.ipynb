## Initial Data Cleaning

```python
# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Setting pandas viewing options
pd.options.display.max_rows=100
pd.options.display.max_columns=100
```

```python
# Importing data as dataframe
df = pd.read_csv('../datasets/train.csv')
```

```python
# In the following cells I take a high-level look at the data.
# This is done with careful reference to the data dictionary, here: http://jse.amstat.org/v19n3/decock/DataDocumentation.txt
# In particular, I'm looking for anything that can be removed outright, which would simplify the analysis.
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
# formatting feature names
df.columns = df.columns.str.lower().str.replace(" ","_")
```

```python
df.info()
```

```python
df.isnull().sum()[59:]
```

```python
# Several features contain mostly nulls and so imputation would cause too great a distortion.
# They are removed here:
df.drop(columns=['alley','fireplace_qu','pool_qc','fence','misc_feature'], inplace=True)
```

```python
df[['lot_frontage','lot_area']].corr()
```

```python
# In addition, lot_frontage contains 300 missing values and is positively correlated with total lot area anyway.
# Will also remove this feature:
df.drop(columns=['lot_frontage'], inplace=True)
```

```python
df.isnull().sum()
```

```python
# garage and basement both have a fair number of null values and, as well, there are several features devoted to each of these.
# Most of these features appear redundant.
# Here, I reduce the number of features so there is one oridinal variable each for garage and basement, expressing overall quality
df.drop(columns=['bsmt_qual',
                 'bsmt_cond',
                 'bsmt_exposure',
                 'bsmtfin_sf_1',
                 'bsmtfin_type_2',
                 'bsmtfin_sf_2',
                 'bsmt_unf_sf',
                 'garage_yr_blt',
                 'garage_finish',
                'garage_cars',
                'garage_cond'],
       inplace= True)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
# For remaining garage and basement variables, NaN indicates no garage or basement. 
# These values can be numeritized, letting 0 indicate no basement/garage.
# Remapping of non-null values is done here:
basement_fin_remap = {'GLQ': 6,
                      'ALQ':5,
                      'BLQ':4,
                      'Rec':3,
                      'LwQ':2,
                      'Unf':1
                     }

garage_qual_remap = {'Ex':5,
                      'Gd':4,
                      'TA':3,
                      'Fa':2,
                      'Po':1
                    }

df['bsmtfin_type_1'] = df['bsmtfin_type_1'].map(basement_fin_remap)
df['garage_qual'] = df['garage_qual'].map(garage_qual_remap)
```

```python
print(df['bsmtfin_type_1'].unique())
print(df['garage_qual'].unique())
```

```python
# Null values are replaced with appropriate filler:

df['bsmtfin_type_1'].fillna(0,inplace=True)
df['garage_qual'].fillna(0,inplace=True)
df['garage_type'].fillna("none",inplace=True)

# Converted to int
df['bsmtfin_type_1'] = df['bsmtfin_type_1'].astype('int')
df['garage_qual'] = df['garage_qual'].astype('int')
```

```python
print(df['bsmtfin_type_1'].unique())
print(df['garage_qual'].unique())
print(df['garage_type'].unique())
```

```python
df.isnull().sum()
```

```python
# There are now a very small number of non-null values, which can be removed:
df.dropna(inplace=True)

df.isnull().sum()
```
