```python
import pandas as pd
import numpy as np

pd.options.display.max_rows=100
pd.options.display.max_columns=100
```

```python
df = pd.read_csv('../datasets/test.csv')
```

```python
df
```

```python
df.shape
```

```python
df.columns = df.columns.str.lower().str.replace(" ","_")
```

```python
df.drop(columns=['alley','fireplace_qu','pool_qc','fence','misc_feature','lot_frontage'], inplace=True)
```

```python
df.drop(columns=['bsmt_qual',
                 'bsmt_cond',
                 'bsmt_exposure',
                 'bsmtfin_sf_1',
                 'bsmtfin_type_2',
                 'bsmtfin_sf_2',
                 'bsmt_unf_sf',
                 'garage_yr_blt',
                 'garage_finish',
                'garage_cars',
                'garage_cond'],
       inplace= True)
```

```python
basement_fin_remap = {'GLQ': 6,
                      'ALQ':5,
                      'BLQ':4,
                      'Rec':3,
                      'LwQ':2,
                      'Unf':1
                     }

garage_qual_remap = {'Ex':5,
                      'Gd':4,
                      'TA':3,
                      'Fa':2,
                      'Po':1
                    }

df['bsmtfin_type_1'] = df['bsmtfin_type_1'].map(basement_fin_remap)
df['garage_qual'] = df['garage_qual'].map(garage_qual_remap)
```

```python
print(df['bsmtfin_type_1'].unique())
print(df['garage_qual'].unique())
```

```python
# Null values are replaced with appropriate filler:

df['bsmtfin_type_1'].fillna(0,inplace=True)
df['garage_qual'].fillna(0,inplace=True)
df['garage_type'].fillna("none",inplace=True)

# Converted to int
df['bsmtfin_type_1'] = df['bsmtfin_type_1'].astype('int')
df['garage_qual'] = df['garage_qual'].astype('int')
```

```python
print(df['bsmtfin_type_1'].unique())
print(df['garage_qual'].unique())
print(df['garage_type'].unique())
```

```python
df.isnull().sum()
```

```python
df['mas_vnr_type'].value_counts()
```

```python
df['mas_vnr_type'].fillna('None', inplace= True)
```

```python
df['mas_vnr_area'].fillna(0, inplace= True)
```

```python
df['electrical'].value_counts()
```

```python
df['electrical'].fillna('SBrkr', inplace= True)
```

```python
df.dtypes
```

```python
df['age'] = df['yr_sold'] - df['year_built']
df['years_reno'] = df['yr_sold'] - df['year_remod/add']
```

```python
df[['bsmtfin_type_1','garage_qual','overall_qual']] = df[['bsmtfin_type_1','garage_qual', 'overall_qual']].astype('str')
```

```python
df.dtypes
```

```python
df['gr_liv_area*total_basement_sf'] = df['gr_liv_area'] * df['total_bsmt_sf']
df['age*years_reno'] = df['gr_liv_area'] * df['total_bsmt_sf']
df['gr_liv_area**2'] = df['gr_liv_area'] ** 2
```
